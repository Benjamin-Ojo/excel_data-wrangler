```python
# Import packages. 
import pandas as pd 
import os 

# Excel packages.
from openpyxl import Workbook as wb 
from openpyxl import load_workbook as lwb
```

```python
# loaing packages. 
xl = pd.read_excel('extract.xlsx', sheet_name = '196',engine='openpyxl', na_values=['nan'])

# View file
xl
```

```python
code = xl.iloc[1, :].dropna().tolist()

code
```

```python
# Deleting rows
xl.drop([0,1], axis=0, inplace=True)

# checking file
xl
```

```python
# Expected Columns. 
columns = ['S/N','DESCRIPTION','CODE','QUANTITY','UNIT PRICE IN EURO','TOTAL PRICE IN EURO ','EXCHANGE_RATE',
              'UNIT PRICE IN NAIRA','TOTAL PRICE IN NAIRA','CLIENT ','INVOICE NUMBER','INVOICE VALUE']

# checking to see the number of columns. 
if len(xl.columns) != len(columns):
    xl[xl.iloc[]]

# converting first row to header. 
xl.columns = ['S/N','DESCRIPTION','CODE','QUANTITY','UNIT PRICE IN EURO','TOTAL PRICE IN EURO ','EXCHANGE_RATE',
              'UNIT PRICE IN NAIRA','TOTAL PRICE IN NAIRA','CLIENT ','INVOICE NUMBER','INVOICE VALUE']

# Drop first row
xl.drop([2], axis= 0, inplace = True)
xl.reset_index()

# Checking values.
xl 
```

```python
# Creating columns.
xl['SUPPLIER_NAME'] = ' '
xl['CONTAINER_CODE'] = code[0]

# Checking file
xl
```

```python
# Reordering columns 
xl = xl.loc[:, ['S/N', 'CONTAINER_CODE', 'SUPPLIER_NAME','DESCRIPTION', 'CODE', 'QUANTITY', 
     'UNIT PRICE IN EURO','TOTAL PRICE IN EURO ', 'EXCHANGE_RATE', 'UNIT PRICE IN NAIRA',
     'TOTAL PRICE IN NAIRA', 'CLIENT ', 'INVOICE NUMBER', 'INVOICE VALUE']]

# Checking columns order
xl.columns
```

```python
# Drop columns that aren't needed
del xl['S/N']
del xl['INVOICE VALUE']

# Check 
xl
```

```python
# Supplier and item code list 
supplier = xl['DESCRIPTION'].fillna('re').to_list()
item_code = xl['CODE'].fillna('re').to_list()

# Check
supplier, item_code
```

```python
supplier_name = []
for i in range(len(supplier)):
    if supplier[i] != 're' and item_code[i] == 're':
        supplier_name.append(supplier[i])
    elif supplier[i] == 're' and item_code[i] == 're':
        supplier_name.append('delete_row')
    else:
        cont=supplier_name[-1].replace('~', '')
        supplier_name.append(f'{cont}~')
```

```python
 # Checking values. 
supplier_name
```

```python
# Defining supplier name
xl['SUPPLIER_NAME'] = supplier_name

xl 
```

```python
# Rows to Delete.
delete_row = xl[xl['SUPPLIER_NAME'] == 'delete_row'].index.to_list()
```

```python
# Drop rows. 
xl.drop(delete_row, axis = 0, inplace = True)

# Reset index. 
xl = xl.reset_index(drop = True)
```
